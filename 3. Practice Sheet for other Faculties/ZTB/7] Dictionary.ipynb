Dictionary is a data structure to hold a collection of values in a single variable
Python dictionary represents a mapping between a key and a value.
In simple terms, it can store pairs of keys and values. 

Each key is linked to a specific value.
-> Can have any number of items with a unique key
-> Dynamic: don't have to specify ANY FIXED SIZE, can easily add/remove elements

A dictionary is MUTABLE (can change individual items)

Popularly known as map/hash table/associative arrays in other langs

Each item in the dictionary has a key:value pair

```python
#Creating Empty Dictionary

create_dict = {}
print(create_dict)
```

```python
dictionary = {'Name': 'Bob', 'Profession':'Builder', 'Age':35}
print(dictionary)

#All items enclosed between curly braces seperated by commas
#Colon(:) used to seperate key from its value
#Keys are unique and can be immutable data type(strings/numbers/tuples)
#Values don't have to be unique & can be of any data type including dictionaries
```

```python
#Creating dictionary using dict()

#keys = immutable values = string, numbers/integers/float, tuple
#value = mutable/immutable = string, numbers/integers, tuple, list, float

#Integer keys
create_dict = dict( {1:['cat',12,12.3], 2:'dog', 3:'python'} )
print(create_dict)

#String keys
create_dict = dict( {'1':'cat', '2':'dog', '3':'python'} )
print(create_dict)

#Tuple keys
create_dict = dict( {(1,2):'cat', (3,4):'dog', (5,6):'python'} )
print(create_dict)

#List key (mutable) = Error
# create_dict = dict( {[1,2]:'cat', [3,4]:'dog', [5,6]:'python'} )
# print(create_dict)
#UNHASHABLE TYPE LIST AS LIST=MUTABLE

#Float datatype keys
create_dict = dict( {1.2:'cat', 2.1:'dog', 3.5:'python'} )
print(create_dict)
```

```python
#Dictionary with mixed keys

create_dict = {'Courses':['CSE110','CSE111'], 
               1:{10,20,15}, 
               (1,2):'A', 
               2.1:1 }
print(create_dict)
```

```python
#Nested Dictionary

create_dict = {'Course':'CSE110', 'highest':{'name':'Troy', 'mark':50}}
print(create_dict)
```

```python
#ACCESSING
#Just like strings/lists, 
#can use square brackets but keys must be used instead of index
#KEYERROR IF KEY NOT FOUND

#Can use built-in function get(key,value)
#-> KEY: TO BE SEARCHED
#-> VALUE: OPTIONAL: If no value found, the set value will be returned. 
#If not set, will return "None".

#Keys must be any available key in the dictionary

capt = {'Portugal':'Ronaldo', 'Argentina':'Messi', 'Spain':'Ramos'}

#KEY : VALUE
#Portugal : RONALDO
#Argentina : MESSI
#SPAIN : RAMOS

print(capt['Portugal']) #Ronaldo
# print(capt[' Brazil']) #KeyError
#so get method returns smth; either none or set value
#square bracketing [] gives error if value not found
print(capt.get('Argentina')) #Messi 
print(capt.get('Argentina', 'Not found')) #Messi 
print(capt.get('Brazil')) #None as not a valid key
print(capt.get('Brazil','Not found')) 
#Not found returned (not None) if not valid key
```

```python
#Dictionary inside a dictionary (nested)
#Use [][]

capt = {'A':'1', 'B':'2', 'C':{'D':'3', 'E':'4', 'F':'5'}, 'G':6}
print(capt['C']) #'D:''3', 'E':'4', 'F':'5'
print(capt['C']['E']) #'4'
print(capt['C']['F']) #'5'


capt_ = {'A':'1', 'B':'2', 'C':{'D':'3', 
                                'E':{'Z':'4', 'X':'10', 
                                     'Y':{'M':'11', 'N':'100'}, 
                                     'F':'5'}, 'G':6}}
print(capt_['C']['E']['Y']['N'])  #100
         
```

```python
#Dictionaries are mutable/changeable
#Like lists, key:value can be changed easily

#change, add, delete

#Variable[Key] = value
#VARIABLE = EXISTING DICTIONARY NAME
#KEY = NEW/EXISTING KEY
#VALUE = NEW VALUE

capt_ = {'A':'1', 'B':'2', 'C':{'D':'3', 
                                'E':{'Z':'4', 'X':'10', 
                                              'Y':{'M':'11', 'N':'100'}, 
                                              'F':'5'}, 'G':6}}
print(capt_)
print(capt_['C']['E']['Y']['N'])  #100
print("=====================================================================================================")

#UPDATE/CHANGE
capt_['C']['E']['Y']['N'] = 200
print(capt_['C']['E']['Y']['N'])  #200
print(capt_)

#ADD NEW VALUE
capt_['Spain']='Casillas'
print(capt_)
print("=====================================================================================================")

capt_['C']['E']['Y']['Argentina']='Messi'
print(capt_)
print("=====================================================================================================")
```

```python
#Continue adding / updating elements using BUILT IN function

#update(dictionary)=>> 
#If key is present, updates. 
#Else, adds key:value pair

book = {'author':'Enid Blyton', 
        'novel':'Famous five', 
        'year':'1998'}
print(book)
print("=========================")

print("updates novel key")
book.update({'novel':'The Secret Seven'}) #updates novel key
print(book)
print("=========================")
book.update(novel='The Secret Seven') #updates novel key
print(book)
print("=========================")

print("adds series key")
book.update({'series':'Five goes on Adventure'}) #adds series key
print(book)
print("=========================")
book.update(series='Five goes on Adventure') #adds series key
print(book)
print("=========================")

```

```python
#Popitem

book = {'author':'Enid Blyton', 'novel':'Famous five', 'year':'1998'}
print(book)
remove = book.popitem() #removes last item
print(remove)
print(book)
```

```python
#Pop

book = {'author':'Enid Blyton', 'novel':'Famous five', 'year':'1998'}
print(book)
remove = book.pop('novel') #pop the key; don't give keys that don't exist 
print(remove)
print(book)
```

```python
book = {'author':'Enid Blyton', 'novel':'Famous five', 'year':'1998'}

#Doesn't take any parameter.
#Returns a view obj that contains a list of all the keys
list_keys = book.keys()
print(list_keys)

#Doesn't take any parameter.
#Returns a view obj that contains a list of all the values
list_values = book.values()
print(list_values)

#Doesn't take any parameter.
#Returns a view obj that contains 
#a list of all the key:value pairs in TUPLE form
list_items_tuple = book.items()
print(list_items_tuple)

# [('author', 'Enid Blyton'), ('novel', 'Famous five'), ('year', '1998')]
```

```python
#IMPORTANT

book = {'author':'Enid Blyton', 
        'novel':'Famous five', 
        'year':'1998'}

#iterate over keys method 1:
for i in book:
    print(i)

print("=========================")
    
#iterate over keys method 2:
# ['author', 'novel', 'year'] = x = book.keys()
for key in book.keys(): 
    #list of all the keys er upor iterating
    print(key)
    
print("=========================")

#iterate over values 
for value in book.values(): #list of all the values er upor iterating
    print(value)

# for value in book: #Trick question, value here is actually the count. 
# NOT value of key.
#     print(value)

print("=========================")

#iterate over key:value
for key,value in book.items(): #iterating key, value simultaneously
    print(key + ' is ' + value)

print("=========================")

#iterate over key:value
for key in book:
    print(key, 'is', book[key])
```

```python
#built-in functions (len, max, min, sorted)

book = {'author':'Enid Blyton', 
        'novel':'Famous five', 
        'year':'1998', 
        'ZEE':'zee'}

print(len(book))
print(max(book.keys())) #y ascii value
print(max(book.values()))#F ascii value
print(min(book.keys()))
print(min(book.values()))
print(sorted(book.keys())) #according to ascii (ascending)
print(sorted(book.values())) #according to ascii (ascending)
```

```python
# setdefault() method default value can be assigned to a key in the dictionary. 
# In case the key doesn’t exist already, 
# then the key will be inserted into the dictionary, 
# and the value becomes the default value, 
# OTHERWISE, None will be inserted if a value is not mentioned.

# dictionary.setdefault(key, value)
# key doesn't exist, insert key, value=given
# key exists, doesn't change value

# dictionary.setdefault(key) 
# key doesn't exist, insert key, value=None 
# key exists, doesn't change value

person_details = {"name": "Jessa", "country": "USA", "telephone": 1178}

# set default value if key doesn't exists
person_details.setdefault('state', 'Texas')

# key doesn't exists and value not mentioned. default None
person_details.setdefault("zip")

# key exists and value mentioned. doesn't  change value
person_details.setdefault('country', 'Canada')

# key exists and value mentioned. doesn't  change value
person_details.setdefault('country')

# Display dictionary
for key, value in person_details.items():
    print(key, ':', value)
```

```python
#JOIN 2 dictionaries
# the dictionary to be added will be passed 
# as the argument to the update() method 
# and the updated dictionary will have items of both the dictionaries.

dict1 = {'Jessa': 70, 'Arul': 80, 'Emma': 55}
dict2 = {'Kelly': 68, 'Harry': 50, 'Olivia': 66}

# copy second dictionary into first dictionary
dict1.update(dict2)
# printing the updated dictionary
print(dict1)

#ALTERNATE

dict3 = {**dict1, **dict2}
print(dict3)
```

```python
#What happens if the second dictionary has a key 
#that already exists in the first dictionary when joining?

dict1 = {'Jessa': 70, 'Arul': 80, 'Emma': 55}
dict2 = {'Kelly': 68, 'Harry': 50, 'Emma': 66}

# join two dictionaries with some common items
dict1.update(dict2)
# printing the updated dictionary
print(dict1)
print(dict1['Emma'])

#ALTERNATE

dict3 = {**dict1, **dict2}
print(dict3)
print(dict3['Emma'])

#Is the below possible?

dict4 = dict1 + dict2
print(dict4)
```

```python
dict1 = {'Jessa': 70, 'Emma': 55}

# Copy dictionary using copy() method
dict2 = dict1.copy()
# printing the new  dictionary
print(dict2)
# output {'Jessa': 70, 'Emma': 55}

# Copy dictionary using dict() 
dict3 = dict(dict1)
print(dict3)
# output {'Jessa': 70, 'Emma': 55}

# Copy dictionary using the output of items() methods
dict4 = dict(dict1.items())
print(dict4)
# output {'Jessa': 70, 'Emma': 55}

# Copy dictionary using the = operator 
dict1 = {'Jessa': 70, 'Emma': 55}

# Copy dictionary using assignment = operator
dict2 = dict1

# modify dict2
dict2.update({'Jessa': 90})
print(dict2)
# Output {'Jessa': 90, 'Emma': 55}

print(dict1)
# Output {'Jessa': 90, 'Emma': 55}
```
